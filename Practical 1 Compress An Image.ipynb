```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
im = cv2.imread('lena.tif',0)
np.shape(im)
x,y = np.shape(im)
X = x//3
Y =y//3
print(X,Y)
plt.imshow(im)
plt.show()
```

```python
im1 = np.zeros((X,Y), np.float32)
row = 0
li = []
col = 0
for i in range(0,x-3,3):
    col = 0
    for j in range(0,y-3,3):
        sumit=0
        mv=0
        for k in range(i,i+3):
            for l in range(j,j+3):
                mv +=(im[k,l]/9)
        li.append(mv)
        im1[row,col] = mv
        col+=1
    row +=1
print(len(li))
plt.imshow(im1)
plt.show()
```
